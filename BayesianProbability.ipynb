### Conditional Probability

The probability of $A$ given $B$: $P(A|B)$

Note the for independent probabilities: $P(A|B) = P(A)$

The foregoing amends the product and sum rules...

$P(A,B) = P(A) \times P(B|A)$

$P(A\ \mathrm{OR}\ B) = P(A) + P(B) - P(A) \times P(B|A)$

### Bayes Theorem

Bayes theorem allows us to reverse conditional probabilites in order to gain a better understanding of what we area actually coditioning on. THat is, if we known the conditional probability of an observation given our belief, we are able to determine the strength of that belief given the observation.

$P(\mathrm{belief}|\mathrm{data}) = \frac{P(\mathrm{data}|\mathrm{belief})\times P(\mathrm{belief})}{P(\mathrm{data})}$

In general:

$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$, where

posterior: $P(A|B)$

prior: $P(A)$ - Often controversial as considered subjective.

likelihood: $P(B|A)$

normalisation: $P(B)$ - May be omitted in the case a ratio of probabilities is all that is necessary.

$\mathrm{Beta}(\alpha_\mathrm{posterior},\beta_\mathrm{posterior})
= \mathrm{Beta}(\alpha_\mathrm{likelihood}+\alpha_\mathrm{prior},
\beta_\mathrm{likelihood}+\beta_\mathrm{prior})$


```python
a = [2, 50]
b = [10, 50]

x = np.arange(0,1,0.00001)
dist1 = beta(a[0],b[0])
dist2 = beta(a[0]+a[1],b[0]+b[1])

plt.plot(x, dist1.pdf(x), label='Beta: likelihood')
plt.plot(x, dist2.pdf(x), label='Beta: likelihood + prior')
plt.title('Beta Probability Density Function')
plt.xlabel('Probability')
plt.ylabel('Density')
plt.legend()
plt.show()
```
